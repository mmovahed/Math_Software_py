# Solving Differential Equations with SciPy

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

## Example 1: Ordinary Differential Equation (ODE)

### Define the differential equation

$dy/dt = y$


```python
def dydt(t, y):
    return y
```

### Initial condition

```python
y0 = [1]
```

### Time span


```python
t_span = (0, 5)
t_eval = np.linspace(t_span[0], t_span[1], 100)
```

### Solve the ODE


```python
solution = solve_ivp(dydt, t_span, y0, t_eval=t_eval)
```

### Plot the solution

```python
plt.figure(figsize=(8, 4))
plt.plot(solution.t, solution.y[0], label='y(t)')
plt.title('Solution of dy/dt = y')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Example 2: System of ODEs

$ dx/dt = x - y,\; dy/dt = x + y$

```python
def system(t, z):
    x, y = z
    dxdt = x - y
    dydt = x + y
    return [dxdt, dydt]
```

### Initial condition

```python
z0 = [1, 0]  
```

### Solve system of ODEs

```python
solution_sys = solve_ivp(system, (0, 5), z0, t_eval=t_eval)
```

### Plot the solution

```python

plt.figure(figsize=(8, 4))
plt.plot(solution_sys.t, solution_sys.y[0], label='x(t)')
plt.plot(solution_sys.t, solution_sys.y[1], label='y(t)')
plt.title('Solution of System of ODEs')
plt.xlabel('t')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```
